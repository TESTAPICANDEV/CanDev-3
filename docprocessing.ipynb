```python
from nltk import word_tokenize
import os
import PyPDF2
from gensim.models.doc2vec import Doc2Vec, TaggedDocument 
import re
```

```python
pdfFileNames = [pdf for pdf in os.listdir('./data') if('.pdf' in pdf)]
def pdf2Text(filePath):
    pdfFileObj = open('./data/' + filePath, 'rb')
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj,strict = False)
    
    text = ''
    for i in range(pdfReader.getNumPages()):
        text += pdfReader.getPage(i).extractText()
    return text
def preprocess(text):
    text = text.replace('\n', '')
    return text
```

```python
documents = []
for pdf in pdfFileNames:
    try:
        doc = preprocess(pdf2Text(pdf))
        if(re.search('\s',doc)!=None):
            documents.append(doc)
    
    except:
        print("Goddamn")
```

```python
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), tags=[str(i)]) for i, _d in enumerate(documents)]
```

```python
len(tagged_data)
```

```python

max_epochs = 100
vec_size = 500
alpha = 0.025

model = Doc2Vec(vector_size = vec_size,
               alpha = alpha,
               min_alpha = 0.00025,
               min_count = 1,
               dm = 1)
model.build_vocab(tagged_data)
model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=max_epochs)
    
model.save("DocVec.model")
```

```python
# model = model= Doc2Vec.load("DocVec.model")
vec = model.infer_vector(word_tokenize(documents[0]))
#model.most_similar(vec)
vec
```

```python
vectors = [model.infer_vector(word_tokenize(doc)) for doc in documents ]
```
